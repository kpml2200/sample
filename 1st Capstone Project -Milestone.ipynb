### Custom Method to Define Week Number

```python
# create a custom method to define week number 
def get_week(day):
    if day < 8:
        return 1
    elif day < 15:
        return 2
    elif day < 22:
        return 3
    else:
        return 4
```

### Add Features - Year, Month, Day, Week, Weekday to df_common_brands 

```python
# map function on DateTime column to extract year :Year
df_common_brands['Year']=df_common_brands['DateTime'].map(lambda x: x.year)
```

```python
# map function on DateTime column to extract month : Month 
df_common_brands['Month']=df_common_brands['DateTime'].map(lambda x: x.month)
```

```python
# map function on DateTime column to extract day of the month : Day 
df_common_brands['Day']=df_common_brands['DateTime'].map(lambda x: x.day)
```

```python
# call custom method to exact week  : Week
df_common_brands['Week']=df_common_brands['DateTime'].map(lambda x: get_week(x.day))
```

```python
# map function on DateTime column to exact weekday and apply formatting style to display day of week : Weekday
df_common_brands['Weekday']=df_common_brands['DateTime'].map(lambda x: x.strftime("%A"))
```

```python
# define function for Emperical cumulative density function
def ecdf(data):
    
    # no of data points
    n=len(data)
    
    # sorted data for x
    x=np.sort(data)
    
    # y-data for the ECDF: y
    y = np.arange(1, n+1) / n

    return x, y
```

```python
# Create an ECDF from real data: x, y
x, y = ecdf(df_common_brands['Purchaseamount'])

# Create a CDF from theoretical samples: x_theor, y_theor
x_theor, y_theor = ecdf(samples)

# Overlay the plots
plt.plot(x_theor, y_theor)
plt.plot(x, y, marker='.', linestyle='none')

# Margins and axis labels
plt.margins(0.02)
plt.xlabel('Purchaseamount')
plt.ylabel('CDF')

# Show the plot
plt.show()
```

```python
def bootstrap_replicate_1d(data, func):
    return func(np.random.choice(data, size=len(data)))

def draw_bs_reps(data, func, size=1):
    """Draw bootstrap replicates."""

    # Initialize array of replicates: bs_replicates
    bs_replicates = np.empty(size)

    # Generate replicates
    for i in range(size):
        bs_replicates[i] = bootstrap_replicate_1d(data, func)

    return bs_replicates
```

```python
#compute a bootstrap estimate of the probability density function of the mean purchase amount

# Take 10,000 bootstrap replicates of the mean: bs_replicates
bs_replicates = draw_bs_reps(df_common_brands.Purchaseamount, np.mean, size=10000)

# Compute and print SEM
sem = np.std(df_common_brands.Purchaseamount) / np.sqrt(len(df_common_brands.Purchaseamount))
print(sem)

# Compute and print standard deviation of bootstrap replicates
bs_std = np.std(bs_replicates)
print("Std :",bs_std)
bs_mean=np.mean(bs_replicates)
print("Mean",bs_mean)

# Make a histogram of the results
_ = plt.hist(bs_replicates, bins=50, normed=True)
_ = plt.xlabel('mean purchase amount in USD')
_ = plt.ylabel('PDF')

# Show the plot
plt.show()
```
